```python
delays = pd.read_csv('delays.csv', header=-1).__array__().T
lambda2 = pd.read_csv('lambda2.csv', header=-1).__array__()

delays = delays[:,2:]
lambda2 = lambda2[:,2:]
```

```python
q2 = 25
```

```python
import matplotlib.pyplot as plt

%matplotlib notebook

plt.figure()
for i,d in enumerate(delays):
    plt.plot(d,((25-25/(1+lambda2)) + q2*d)[0], label='{} fps'.format((i+1)*30))
    
plt.legend()
plt.xlabel('Delay (frames)')
plt.ylabel('MSE')
plt.title('MSE predictor objective with motion covariance = 0.075')
```
